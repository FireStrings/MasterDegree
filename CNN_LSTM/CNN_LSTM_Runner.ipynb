```python
if op == "results":
  list_df_results = []
  for estacao in lista_estacoes:
      results_cnn_lstm = manage_results(_dir+"/last_results/" + estacao + "_lstm.pkl", "read")
      results_cnn_bilstm = manage_results(_dir+"/last_results/" + estacao + "_bilstm.pkl", "read")

      list_df_results.append(create_table_results(estacao, "lstm", results_cnn_lstm[estacao]))
      list_df_results.append(create_table_results(estacao, "bilstm", results_cnn_bilstm[estacao]))

  df_final = pd.concat(list_df_results, axis=0)

elif op == "train":
  for estacao in lista_estacoes:
      anual_df = load_and_filter(estacao, 7, 18).set_index("data_hora")

      results_cnn_lstm[estacao] = {}
      results_cnn_bilstm[estacao] = {}

      # LSTM
      process_cnn_lstm(1, estacao, anual_df, results_cnn_lstm, False, False)
      manage_results(_dir+"/last_results/" + estacao + "_lstm.pkl", "write", results_cnn_lstm)

      # BiLSTM
      process_cnn_lstm(5, estacao, anual_df, results_cnn_bilstm, False, True)
      manage_results(_dir+"/last_results/" + estacao + "_bilstm.pkl", "write", results_cnn_bilstm)
```
